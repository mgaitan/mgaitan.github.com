## Por dónde recortamos ?

Como vemos, el *layout* es muy simétrico y todas las páginas son iguales (4 x 3 escuelas por página).  Lo que queremos es obtener el nombre de la escuela y la fecha y hora del registro de su escrutinio. Por ejemplo `CENTRO EDUC.NIVEL MEDIO ADULTO  27/10 a las 20:40`. Para extraer texto de un pdf, una herramienta sencilla y potente es [pdftotext](http://en.wikipedia.org/wiki/Pdftotext)

```python
!pdftotext
```

La idea es definir un rectangulo con cuyo vértice superior izquierdo se define por `-x -y`  y su ancho y alto por `-W -H`. 
Luego de probar un poquito, encontré el primero en 

```python
!pdftotext -f 1 -l 1 -x 205 -y 140 -W 70 -H 35 C04001.pdf && cat C04001.txt
```

Para la siguiente columna, hay que correr un poco la máscara de cropping sobre el eje X. Me dió una separación de 90 puntos 

```python
!pdftotext -f 1 -l 1 -x 295 -y 140 -W 70 -H 35 C04001.pdf && cat C04001.txt
```

Moviendo el eje `y` convertimos la siguiente fila. El delta vertical me dio 230

```python
!pdftotext -f 1 -l 1 -x 475 -y 370 -W 70 -H 35 C04001.pdf && cat C04001.txt
```

Entonces podemos obtener todos los `(x, y)` que necesitamos


```python
from itertools import product
inicio_x = 205
inicio_y = 140
delta_x = 90
delta_y = 230
X = [inicio_x + delta_x * i for i in range(4)] 
Y = [inicio_y + delta_y * j for j in range(3)]
list(product(X, Y))
```

## Limpiando la salida

La salida de pdftotext se puede capturar, pero hay que limpiarla un poco

```python
!!pdftotext -f 1 -l 1 -x 295 -y 140 -W 70 -H 35 C04001.pdf && cat C04001.txt
```

```python
def clean(capture):
    token = capture.index('')
    return ' '.join(capture[:token]), capture[token + 1]

print clean(['ESC NUESTRA', 'SE\xc3\x91ORA DEL HUERTO', '', '27/10 a las 21:20', '', '\x0c'])
print clean(['NACIONAL DEAN FUNES', '', 'fecha'])
```

¡Pero qué detalle! Eso es sólo para la página 1 (limitado por los flags `-f 1 -l 1`). Pdftotext sabe recortar todas las páginas a la vez

```python
!pdftotext -x 295 -y 140 -W 70 -H 35 C04001.pdf && head C04001.txt 
```

... pero ya limpiamos la captura para **una** escuela y somos vagos, no?. Mejor hagamoslo página a página y listo. 

## ¿Cuantás páginas tiene un PDF?

Google conoce el código que tiene la respuesta

```python
# from http://code.activestate.com/recipes/496837-count-pdf-pages/
import re

rxcountpages = re.compile(r"/Type\s*/Page([^s]|$)", re.MULTILINE|re.DOTALL)

def count_pdf_pages(filename):
    data = file(filename,"rb").read()
    return len(rxcountpages.findall(data))


count_pdf_pages('C04001.pdf')
```

## Listo, ensamblemos los legos

Ejecutamos `pdftotext` para cada región que queremos en cada página de cada pdf descargado. Los resultados limpios, los guardamos en una lista como tuplas `(ID MUNICIPIO, ESCUELA, HORA)`

```python
table = []
for i in range(1, 27):
    filename = "C040%02d" % i
    for x, y in product(X, Y):
        num_pages = count_pdf_pages("%s.pdf" % filename)
        for page in range(1, num_pages + 1):
            extracto = !pdftotext -f {page} -l {page} -x {x} -y {y} -W 70 -H 35 {filename}.pdf && cat {filename}.txt
            try:
                escuela, hora = clean(extracto)
                table.append((i, escuela, hora))
            except ValueError:
                continue
```
